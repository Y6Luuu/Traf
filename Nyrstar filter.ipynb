```python
import pandas as pd
pd.reset_option('display.float_format')
df = pd.read_excel(r"K:\FINANCE\Concs\_ Yuyan\BB Report from Gab\BB report for bank 8-Mar.xlsx", sheet_name = 'Sales Receivables ')
```

```python
#价格筛选
def UPR(a):
    return 1500000< a < 2000000

def Assignment(a):
    return a == "685244.11.1"	
```

```python
cols = ['Purchase Assignment Reference', 'Sales Assignment Reference', 'Purchase Commodity', 'Sales Assignment Counterparty', 'Sales Invoice ID', 'Sales Invoice Date', \
    'Sales Invoice Due Date', 'Unpaid Provisional Receivables(USD)']

df_filter = pd.DataFrame(df, columns=cols)

#筛选开始
df_filter = df_filter.loc[df_filter['Unpaid Provisional Receivables(USD)'].apply(UPR)]
df_filter['Unpaid Provisional Receivables(USD)'] = df_filter['Unpaid Provisional Receivables(USD)'].apply(lambda x:format(int(x), ','))
df_filter
```

```python
df_filter2 = pd.DataFrame(df, columns=cols)
df_filter2 = df_filter2.loc[df_filter2['Sales Assignment Reference'].apply(Assignment)]
df_filter2
```
